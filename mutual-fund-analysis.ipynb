```python
import pandas as pd

# setting data dir
dir = "/kaggle/input/mutualsfundsindia/"

# reading the database
data = pd.read_csv(dir + "mutual_funds_india.csv")
top = data[:100]
bottom = data[-100:]
# printing the top 10 rows
display(data.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Scatter plot with id against one year returns
plt.scatter(data['id'], data['return_1yr'])

# Adding Title to the Plot
plt.title("One Year returns Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Mutual Fund ID')
plt.ylabel('Returns in CAGR')
 
plt.show()


# line plot with id against one year returns
plt.plot(data['id'], data['return_1yr'])

# Adding Title to the Plot
plt.title("One Year returns Line Plot")
 
# Setting the X and Y labels
plt.xlabel('Mutual Fund ID')
plt.ylabel('Returns in CAGR')
 
plt.show()


# Scatter plot with id against one year returns
plt.scatter(top['id'], top['return_1yr'])
plt.title("One Year returns Scatter Plot")
plt.xlabel('Mutual Fund ID')
plt.ylabel('Returns in CAGR')
plt.show()

plt.plot(top['id'], top['return_1yr'])
plt.title("One Year returns Line Plot")
plt.xlabel('Mutual Fund ID')
plt.ylabel('Returns in CAGR')
plt.show()

# Scatter plot with id against one year returns
plt.scatter(bottom['id'], bottom['return_1yr'])
plt.title("One Year returns Scatter Plot")
plt.xlabel('Mutual Fund ID')
plt.ylabel('Returns in CAGR')
plt.show()

plt.plot(bottom['id'], bottom['return_1yr'])
plt.title("One Year returns Line Plot")
plt.xlabel('Mutual Fund ID')
plt.ylabel('Returns in CAGR')
plt.show()
```
